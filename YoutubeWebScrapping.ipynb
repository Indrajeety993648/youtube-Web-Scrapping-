```python
from googleapiclient.discovery import build
from sklearn.feature_extraction.text import TfidfVectorizer
import re

def get_video_id(video_url):
    # Extract video ID from the YouTube video URL
    match = re.search(r"(?<=v=)[\w-]+", video_url)
    if match:
        return match.group(0)
    else:
        print("Invalid YouTube video URL.")
        return None

def get_comments(video_url, api_key):
    video_id = get_video_id(video_url)
    if video_id:
        youtube = build('youtube', 'v3', developerKey=api_key)
        response = youtube.commentThreads().list(
            part='snippet',
            videoId=video_id,
            maxResults=100,  # Adjust as needed
            textFormat='plainText'
        ).execute()

        comments = [item['snippet']['topLevelComment']['snippet']['textDisplay'] for item in response['items']]
        return comments
    else:
        return []

def search_comments(comments_data, pattern):
    matching_comments = [comment for comment in comments_data if re.search(pattern, comment)]
    return matching_comments

# Define the YouTube video URL
video_url = 'https://www.youtube.com/watch?v=0dgq4ZkjQCA&list=PL6gx4Cwl9DGCkg2uj3PxUWhMDuTw3VKjM&index=4'

# Define the API key (replace with your own key for production)
api_key = 'AIzaSyB0WPhyOy0SqjHnUTIe_1uUYZVFg5iruLo'

# Get comments
comments_data = get_comments(video_url, api_key)

# Print comments
print("Comments:")
for comment in comments_data:
    print("-", comment)

# Define a regular expression pattern to search for a specific type of comment
pattern = r'indian | hindi'  # Example: match comments containing "awesome", "love", or "helpful"

# Search for comments matching the pattern
matching_comments = search_comments(comments_data, pattern)

# Print matching comments
print("\nComments matching the pattern:")
for comment in matching_comments:
    print("-", comment)

# Convert comments to vector format
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(comments_data)

# Print out the shape of the TF-IDF matrix
print("\nShape of TF-IDF matrix:", tfidf_matrix.shape)
```
